# Correlation Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_path = '../data/SQuAD_train_augmented_DF.h5'
```

```python
df = pd.read_hdf(data_path, 'df_train')
```

```python
context_lengths = df['context'].apply(lambda x: len(x))
question_lengths = df['question'].apply(lambda x: len(x))
answer_lengths = df['text'].apply(lambda x: len(x))
```

```python
length_df = pd.DataFrame({'context': context_lengths, 'question': question_lengths, 'answer':answer_lengths})
```

```python
sns.heatmap(data=length_df.corr(), annot=True)
```

## There are no strong correlations between context, question, and answer lengths
