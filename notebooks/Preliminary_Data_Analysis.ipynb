# Prelimary Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_path = '../data/SQuAD_train_augmented_DF.h5'
```

```python
df = pd.read_hdf(data_path, 'df_train')
```

## Calculate the length of the contexts, questions, and answers

```python
context_lengths = df['context'].apply(lambda x: len(x))
question_lengths = df['question'].apply(lambda x: len(x))
answer_lengths = df['text'].apply(lambda x: len(x))
```

```python
length_df = pd.DataFrame({'context': context_lengths, 'question': question_lengths, 'answer':answer_lengths})
```

```python
mean_context_length = np.mean(context_lengths)
mean_question_length = np.mean(question_lengths)
mean_answer_length = np.mean(answer_lengths)
```

```python
print(f'The mean context length is: {mean_context_length:0.2f}')
print(f'The mean question length is: {mean_question_length:0.2f}')
print(f'The mean answer length is: {mean_answer_length:0.2f}')
```

```python
median_context_length = np.median(context_lengths)
median_question_length = np.median(question_lengths)
median_answer_length = np.median(answer_lengths)
```

```python
print(f'The median context length is: {median_context_length:0.2f}')
print(f'The median question length is: {median_question_length:0.2f}')
print(f'The median answer length is: {median_answer_length:0.2f}')
```

## Plot the distributions

```python
plt.figure()
sns.histplot(context_lengths)
plt.title('Distribution of Context Length')
plt.show()
plt.savefig('../docs/context_length_distribution.png')
```

```python
plt.figure()
sns.histplot(question_lengths)
plt.title('Distribution of Question Length')
plt.show()
plt.savefig('../docs/question_length_distribution.png')
```

```python
plt.figure()
sns.histplot(answer_lengths)
plt.title('Distribution of Answer Length')
plt.show()
plt.savefig('../docs/answer_length_distribution.png')
```
