# Text Analysis on Raw Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data_path = '../data/SQuAD_train_augmented_DF.h5'
```

```python
df = pd.read_hdf(data_path, 'df_train')
```

```python
df['context'].str.split().explode().value_counts().nlargest(50)
```

```python
df['question'].str.split().explode().value_counts().nlargest(50)
```

```python
df['text'].str.split().explode().value_counts().nlargest(50)
```

```python
wc_context = WordCloud(width=400, height=330, max_words=50, colormap="Dark2", include_numbers=True).generate(df['context'].str.cat(sep=' '))
plt.figure(figsize=(10,8))
plt.imshow(wc_context, interpolation='bilinear')
plt.axis("off")
plt.show()
```
